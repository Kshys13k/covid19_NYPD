# Comparison of total cases between Spain and Poland

```python
dfPol_for_plot=dfPoland.dropna(subset=["total_cases"])
dfPol_for_plot=dfPol_for_plot[["date","total_cases","new_cases","total_deaths","new_deaths"]]
dfPol_for_plot
```

```python
dfEsp_for_plot.to_csv("../data/esp_for_plot.csv", index=False)
dfPol_for_plot.to_csv("../data/pol_for_plot.csv",  index=False)
```
